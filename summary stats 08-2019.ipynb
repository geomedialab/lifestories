```python
storydata = gl.glob('EXPORT/stats/agg_su/*.csv')
```

```python
stories = {}

for databit in storydata:
    stories[databit[20:22]] = pd.read_csv(databit, sep=',', encoding='latin-1')
```

```python


#df = pd.DataFrame(,index=['OG','AP','EH','JR'],columns=['ho','ho','ho','ho'])

#stories['og_su'].describe()

#for story in stories:
#    pd.Series(story[2])
if not os.path.exists('sumstats'):
    os.mkdir('sumstats')
else:
    pass


for (k1,story) in stories.items():    
    i = 0
    scales = ['all',2,3,4,5,6]
    
    for scale in scales:
        
        db = story.loc[story['su_num'].notnull()]
        
        if i > 0:
            db = db.loc[db['scale_order']==scale]
        else:
            pass
        
        ssu = db.loc[db['place']!='_end']#spatial story units
        asu = db.loc[db['place']=='_end']#aspatial story units
        
        value_list = {
            '0.0 storyteller' : k1[:2],
            '0.1 scale' : scale,
            
            '1.1 NUMBER OF su' : db['su_num'].count(),
            '1.2 MEAN NUMBER OF MENTIONS within su' : db['mention_freq'].mean(),
            '1.3 MEAN LENGTH of su (minutes)' : db['num_minutes'].mean(),

            '2.1 NUMBER OF ssu' : ssu['su_num'].count(),
            '2.2 MEAN NUMBER OF MENTIONS within ssu' : ssu['mention_freq'].mean(),
            '2.3 MEAN LENGTH of ssu (minutes)' : ssu['num_minutes'].mean(),

            '3.1 NUMBER OF ssu WITHOUT ANY mentions' : ssu.loc[ssu['mention_freq']==0]['su_num'].count(),
            '3.2 MEAN NUMBER OF MENTIONS within ssu WITHOUT ANY mentions' : ssu.loc[ssu['mention_freq']==0]['mention_freq'].mean(),
            '3.3 MEAN LENGTH of ssu WITHOUT ANY mentions (minutes)' : ssu.loc[ssu['mention_freq']==0]['num_minutes'].mean(),

            #'4.1 NUMBER OF ssu WITHOUT ANY SAME mentions' : ssu.loc[ssu['mention_match_freq']==0]['su_num'].count(),
            #'4.2 MEAN NUMBER OF MENTIONS within ssu WITHOUT ANY SAME mentions' : ssu.loc[ssu['mention_match_freq']==0]['mention_freq'].mean(),
            #'4.3 MEAN LENGTH of ssu WITHOUT ANY SAME mentions (minutes)' : ssu.loc[ssu['mention_match_freq']==0]['num_minutes'].mean(),    

            '5.1 NUMBER OF ssu with ONLY SAME mentions' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_freq - ssu.mention_match_freq == 0),'su_num'].count(),
            '5.2 MEAN NUMBER OF mentions within ssu with ONLY SAME mentions' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_freq - ssu.mention_match_freq == 0),'mention_freq'].mean(),
            '5.3 MEAN LENGTH of ssu with ONLY SAME mentions (minutes)' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_freq - ssu.mention_match_freq == 0),'num_minutes'].mean(),

            '6.1 NUMBER OF ssu with AT LEAST 1 SAME mention and AT LEAST 1 other mention' : ssu.loc[((ssu.mention_freq - ssu.mention_match_freq) > 0) & (ssu.mention_match_freq > 0),'su_num'].count(),
            '6.2.1 MEAN NUMBER OF SAME mentions within ssu with AT LEAST 1 SAME mention and AT LEAST 1 other mention' : ssu.loc[((ssu.mention_freq - ssu.mention_match_freq) > 0) & (ssu.mention_match_freq > 0),'mention_match_freq'].mean(),
            '6.2.2 MEAN NUMBER OF OTHER mentions within ssu with AT LEAST 1 SAME mention and AT LEAST 1 other mention' :  (ssu.loc[((ssu.mention_freq - ssu.mention_match_freq) > 0) & (ssu.mention_match_freq > 0),'mention_freq']-ssu['mention_match_freq']).mean(),
            '6.3 MEAN LENGTH of ssu with AT LEAST 1 SAME mention and AT LEAST 1 other mention' : ssu.loc[((ssu.mention_freq - ssu.mention_match_freq) > 0) & (ssu.mention_match_freq > 0),'num_minutes'].mean(),

            '7.1 NUMBER OF ssu with ONLY OTHER mentions' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_match_freq == 0),'su_num'].count(),
            '7.2 MEAN NUMBER OF mentions within ssu with ONLY OTHER mentions' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_match_freq == 0),'mention_freq'].mean(),
            '7.3 MEAN LENGTH of ssu with ONLY OTHER mentions (minutes)' : ssu.loc[(ssu.mention_freq > 0) & (ssu.mention_match_freq == 0),'num_minutes'].mean(),
        }
        if i == 0:
            df0 = pd.DataFrame.from_dict(value_list, orient='index', dtype=None, columns=['value'])
            df1 = df0.sort_index()
        else:
            df2 = pd.DataFrame.from_dict(value_list, orient='index', dtype=None, columns=['value'])
            df1 = pd.merge(df1, df2,left_index=True,right_index=True,suffixes=(' ',' '))
        
        i += 1

    df1.to_csv('sumstats/'+k1[:2]+'.csv',sep=',', index=True, encoding='latin-1')


#'mention_freq'>0 AND 'mention_freq'-'mention_match_freq'==0


```
